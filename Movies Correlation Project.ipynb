```python
# importing the packages i need in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# reading in the data

df = pd.read_csv(r'/Users/Ali/Desktop/movies.csv')
```

```python
# checking the data

df.head()
```

```python
# checking to see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# dropping any rows with null values

df = df.dropna(how='any',axis=0) 
```

```python
# determining data types for the columns

df.dtypes
```

```python
# changing data type of the columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# creating correct year colomn based on released date

#df.drop(columns=['correctyear'])
df['correctyear'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
df.head()
```

```python
# ordering by gross value

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
# setting the options to see all rows of the database

pd.set_option('display.max_rows', None)
```

```python
# dropping any duplicate (I don't drop any here)

df['column name'].drop_duplicates()
```

```python
# scatter plot with budget vs gross to determine the correlation 

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for the movie')
plt.show
```

```python
# plotting budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
# determining what's the actual correlation is (it shows high correlation between budget & gross)

df.corr(method='pearson') #pearson (default), kendall. spearman
```

```python
#creating heatmap using correlation matrix data

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# unstacking correlation

corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
# pairing up the unstacked correlation

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# finding the high correlations

high_corr = sorted_pairs[(sorted_pairs) > 0.5] 
high_corr
```
